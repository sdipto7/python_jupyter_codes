#TF-IDF

```python
import math
import pandas as pd
```

**Let's consider three students' response answer on the question, "What is Supervised Machine Learning ?"**

```python
student_1_response = "Supervised machine learning is training a machine learning model with a labeled dataset. To simply put, a model learns with a dataset where the actual outcome or result are explicitly mentioned."
student_2_response = "Supervised machine learning involves using a labeled dataset to train a machine learning model. To put it another way, a model learns from a dataset in which the actual outcome or conclusion is explicitly stated."
student_3_response = "To train a machine learning model under supervision, you use a labeled dataset. Models are trained on data when the outcome or result is explicitly stated."
```

Split the response answers into words

```python
word_container_1 = student_1_response.split(' ')
word_container_2 = student_2_response.split(' ')
word_container_3 = student_3_response.split(' ')
```

picking up the unique words

```python
word_set = set(word_container_1).union(set(word_container_2).union(set(word_container_3)))
```

```python
word_set
```

creating two dictionaries with the wordSet

```python
word_dictionary_1 = dict.fromkeys(word_set, 0) 
word_dictionary_2 = dict.fromkeys(word_set, 0)
word_dictionary_3 = dict.fromkeys(word_set, 0)
```

adding frequency of unique words from the response answers into the dictionaries

```python
for word in word_container_1:
    word_dictionary_1[word] += 1
    
for word in word_container_2:
    word_dictionary_2[word] += 1
    
for word in word_container_3:
    word_dictionary_3[word] += 1
```

```python
word_dictionary_1
```

```python
word_dictionary_2
```

```python
word_dictionary_3
```

calculate tf values for the dictionary keys values / word counts of the word_containers

```python
def TFValue(dictionary, word_container_count):
    tf_dictionary = {}
    for word, count in dictionary.items():
        tf_dictionary[word] = count/float(word_container_count)
    return tf_dictionary
```

```python
tf_container_1 = TFValue(word_dictionary_1, len(word_container_1))
tf_container_2 = TFValue(word_dictionary_2, len(word_container_2))
tf_container_3 = TFValue(word_dictionary_3, len(word_container_3))
```

```python
tf_container_1
```

calculate idf value for the word lists

```python
def IDFValue(word_list):
    idf_dictionary = {}
    idf_dictionary = dict.fromkeys(word_list[0].keys(), 0)
    
    for doc in word_list:
        for word, val in doc.items():
            if val > 0:
                idf_dictionary[word] += 1
    
    for word, val in idf_dictionary.items():
        idf_dictionary[word] = math.log10(len(word_list) / float(val))
        
    return idf_dictionary
```

```python
idfs = IDFValue([word_dictionary_1, word_dictionary_2, word_dictionary_3])
```

calculate tf * idf

```python
def TF_IDFValue(tf_container, idfs):
    tf_idf = {}
    for word, val in tf_container.items():
        tf_idf[word] = val * idfs[word]
    return tf_idf
```

```python
tf_idf_container_1 = TF_IDFValue(tf_container_1, idfs)
tf_idf_container_2 = TF_IDFValue(tf_container_2, idfs)
tf_idf_container_3 = TF_IDFValue(tf_container_3, idfs)
```

creating a dataframe to understand the tf * idf values

```python
pd.DataFrame([tf_idf_container_1, tf_idf_container_2, tf_idf_container_3])
```
