# Loading the dataset into DataFrame

```python
#Loading dataset as dataframe
df = pd.read_csv('heart failur classification dataset.csv')
df.head()
```

```python
#droping the 'unnamed: 0' column from the data-set
df.drop('Unnamed: 0', axis='columns', inplace=True)
```

# Handling the null values of the features with SimpleImputer
### Depending on the dataset, we would have handled the null values of the label also if the label consists of null values.

```python
#null value check of the features
df.isnull()
```

```python
#Total null values in each feature
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Imputing the serum_sodium column's NaN values with mean strategy
impute1 = SimpleImputer(missing_values = np.nan, strategy = 'mean')

impute1.fit(df[['serum_sodium']])

df['serum_sodium'] = impute1.transform(df[['serum_sodium']])

# Imputing the time column's NaN values with mean strategy
impute2 = SimpleImputer(missing_values = np.nan, strategy = 'mean')

impute2.fit(df[['time']])

df['time'] = impute2.transform(df[['time']])
```

```python
#Checking if the NaN values are filled or not.
df.isnull().sum()
```

```python
df.head()
```

# Selecting the features and labels

```python
# Selecting the features
features = df[df.columns.difference(['DEATH_EVENT'])]
features
```

```python
#selecting the labels
labels = df[['DEATH_EVENT']]
labels
```

# Scaling the features with StandardScaler
### Depending on the dataset, we would have scaled the label also if the label values required scaling.

```python
from sklearn.preprocessing import StandardScaler

#Scaling the feature values with StandardScaler
scaler = StandardScaler()
scaler.fit(features)

features_Standard_scaled = scaler.transform(features)
```

# Scaling the features with MinMaxScaler
### Depending on the dataset, we would have scaled the label also if the label values required scaling.

```python
from sklearn.preprocessing import MinMaxScaler

#Scaling the feature values with MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(features)

features_MinMax_scaled = scaler.transform(features)
```

# Scaling the features with RobustScaler
### Depending on the dataset, we would have scaled the label also if the label values required scaling.

```python
from sklearn.preprocessing import RobustScaler

#Scaling the feature values with RobustScalar
scaler = RobustScaler()
scaler.fit(features)

features_Robust_scaled = scaler.transform(features)
```

# Splitting the Unscaled features and labels into training and testing set
### And, fitting into the KNeighborsClassifier

```python
from sklearn.model_selection import train_test_split

#splitting the dataset into training set and testing set using the scaled features and labels.
X_train, X_test, Y_train, Y_test = train_test_split(features, labels, test_size=0.20, random_state=1)

print(f'X_train shape: {X_train.shape}')
print(f'Y_train shape: {Y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'Y_test shape: {Y_test.shape}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_train, Y_train)

print("Test accuracy without scaling: {:.2f}".format(knn.score(X_test, Y_test)))
```

# Splitting the MinMaxScaled features and labels into training and testing set
### And, fitting into the KNeighborsClassifier

```python
#Using the MinMax_scaled features for training
X_train, X_test, Y_train, Y_test = train_test_split(features_MinMax_scaled, labels, test_size=0.20, random_state=1)

knn.fit(X_train, Y_train)

print("Test accuracy with MinMax scaling: {:.2f}".format(knn.score(X_test, Y_test)))
```

# Splitting the StandardScaled features and labels into training and testing set
### And, fitting into the KNeighborsClassifier

```python
#Using the Standard_scaled features for training
X_train, X_test, Y_train, Y_test = train_test_split(features_Standard_scaled, labels, test_size=0.20, random_state=1)

knn.fit(X_train, Y_train)

print("Test accuracy with Standard scaling: {:.2f}".format(knn.score(X_test, Y_test)))
```

# Splitting the RobustScaled features and labels into training and testing set
### And, fitting into the KNeighborsClassifier

```python
#Using the Robust_scaled features for training
X_train, X_test, Y_train, Y_test = train_test_split(features_Robust_scaled, labels, test_size=0.20, random_state=1)

knn.fit(X_train, Y_train)

print("Test accuracy with Robust scaling: {:.2f}".format(knn.score(X_test, Y_test)))
```
