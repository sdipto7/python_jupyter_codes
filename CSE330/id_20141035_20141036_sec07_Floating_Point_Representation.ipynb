# Floating Point Representation

## 8 bit Floating Point Representation


We will implement a 8 bit Floating point number system in this part

Where First bit is a sign bit, next three bits are exponents\
and the rest 4 bits are mantissa (significand).

Numbers are divided into two categories, normals and subnormals.

For normals the encoding is:    (-1)^sign * 2^(e-3) * 1.fraction\
For subnormals the encoding is: (-1)^sign * 2^(1-3) * 0.fraction

Example:
    0 001 0000 = + 2^(1-3) * 1.0  = +0.25\
    0 010 0000 = + 2^(2-3) * 1.0  = +0.5\
    0 011 0000 = + 2^(3-3) * 1.0  = +1.0\
    0 011 0100 = + 2^(3-3) * 1.25 = +1.25

If exponent is 0 then the number is considered subnormal.
example,
    0 000 0000 = + 2^-2 * 0   = +0.0\
    0 000 1000 = + 2^-2 * 0.5 = +0.125\
    0 000 1100 = + 2^-2 * 0.75 = +0.1875

If exponent is 7 then and mantissa is zero then the number is 
considered infinity (inf), otherwise it is not-a-number (nan).
example
    0 111 0000 = +inf\
    1 111 0000 = -inf\
    X 111 XXXX = nan

```python
import math
import itertools
import matplotlib.pyplot as plt
```

```python
class Float8():
    
    def __init__(self, bitstring):
#         Constructor
#         takes a 8-bit string of 0's and 1's as input and stores the sub-strings accordingly.
        
        assert(len(bitstring)==8)  # Make sure the input consists of exactly 8-bits.
        
        
        self.sign = bitstring[0]       # Assign the sign bit
        self.exponent = bitstring[1:4] # Assign the exponent part
        self.mantissa = bitstring[4:]  # Assign the mantissa

        self.val = self.calculate_value()
        
    def __str__(self):
        return f'Sign bit value: {self.sign}\n' + \
            f'Exponent value: {self.exponent}\n' + \
            f'Mantissa value: {self.mantissa}\n' + \
            f'Floating value: {self.val}\n'
    
    def tobitstring(self):
        return self.sign + self.exponent + self.mantissa
    
    def toformattedstring(self):
        return ' '.join([self.sign, self.exponent, self.mantissa])
    
    def calculate_value(self):
        
        val = 0.0       
        if self.exponent == '111': # Handling the case of infinity and not and a number
            if self.mantissa != "0000":
                val = math.nan
            else:
                if self.sign == '1': 
                    val = -math.inf
                else:
                    val = math.inf
        
        elif self.exponent == '000':  # Handle the case of subnormals  #0 000 0110 
            
            #The mantissa/fraction part
            for index, bit in enumerate(self.mantissa):
                val += (int(bit) * pow(2,-(int(index)+1)))

            ## now the exponent part -> 2^(1-3)
            val *= pow(2,(1-3))
            
            ##now the signpart -> (-1)^sign
            if self.sign == '1':
                val = - val
            
        else:  ##handling the case of normals
            
            #The mantissa/fraction part
            val = 1.0
            for index, bit in enumerate(self.mantissa):
                val += (int(bit) * pow(2,-(int(index)+1)))
                
            ## now the exponent part -> 2^(e-3)
            e = 0
            for index, bit in enumerate(reversed(self.exponent)):
                e += (int(bit)*pow(2,int(index)))
            
            val *= pow(2,(e-3))

            ##now the signpart -> (-1)^sign
            if self.sign == '1':
                val = -val
        
        
        return val

```

# 16 Bit Floating Point Representation


Over here, we will implement a 16 bit Floating point number system.\
Where First bit is a sign bit, next four bits are exponents\
and the rest 11 bits are mantissa (significand).

numbers are divided into two categories, normals and subnormals.

for normals the encoding is:    (-1)^sign * 2^(e-7) * 1.fraction\
for subnormals the encoding is: (-1)^sign * 2^(1-7) * 0.fraction

example:
    0 0101 00000000000 = + 2^(5-7) * 1.0  = +0.25\
    0 0110 00000000000 = + 2^(6-7) * 1.0  = +0.5\
    0 0111 00000000000 = + 2^(3-7) * 1.0  = +1.0\
    0 0111 01000000000 = + 2^(3-7) * 1.25 = +1.25

if exponent is 0 then the number is considered subnormal.\
example,
    0 0000 00000000000 = + 2^-6 * 0    = +0.0\
    0 0000 10000000000 = + 2^-6 * 0.5  = +0.0078125\
    0 0000 11000000000 = + 2^-6 * 0.75 = +0.01171875

if exponent is 15 (all 1's) then and mantissa is zero then the number\
is considered infinity (inf), otherwise it is not-a-number (nan).
example
    0 1111 00000000000 = +inf\
    1 1111 00000000000 = -inf\
    X 1111 XXXXXXXXXXX = nan

```python
class Float16():    

    def __init__(self, bitstring):
#         Constructor
#         takes a 16-bit string of 0's and 1's as input and stores the sub-strings accordingly.

        assert(len(bitstring)==16)      # Make sure the input consists of exactly 16-bits.
        
        self.sign = bitstring[0]        # Assign the sign bit        
        self.exponent = bitstring[1:5]  # Assign the exponent part        
        self.mantissa = bitstring[5:]   # Assign the mantissa     

        self.val = self.calculate_value()
        
    def __str__(self):
        return f'Sign bit value: {self.sign}\n' + \
            f'Exponent value: {self.exponent}\n' + \
            f'Mantissa value: {self.mantissa}\n' + \
            f'Floating value: {self.val}\n'
    
    def tobitstring(self):
        return self.sign + self.exponent + self.mantissa
    
    def toformattedstring(self):
        return ' '.join([self.sign, self.exponent, self.mantissa])
    
    def calculate_value(self):
        
        val = 0.0
        if self.exponent == '1111':     # Handling the case of inf and nan
            if self.mantissa != "00000000000":
                val = math.nan
            else:
                if self.sign == '1': 
                    val = -math.inf
                else:
                    val = math.inf
                    
        elif self.exponent == '0000':   # Handling the case of subnormals
            
            #The mantissa/fraction part
            for index, bit in enumerate(self.mantissa):
                val += (int(bit) * pow(2,-(int(index)+1)))
                
            ## now the exponent part -> 2^(1-7)
            val *= pow(2,(1-7))
            
            ##now the signpart -> (-1)^sign
            if self.sign == '1':
                val = (-1) * val
            
        else:  # Handling the case of normals
            
            #The mantissa/fraction part
            val = 1.0
            for index, bit in enumerate(self.mantissa):
                val += (int(bit) * pow(2,-(int(index)+1)))
                
            ## now the exponent part -> 2^(e-7)  #1011
            e = 0
            for index, bit in enumerate(reversed(self.exponent)):
                e += int(bit)*pow(2,int(index))
            
            val *= pow(2,(e-7))

            ##now the signpart -> (-1)^sign
            if self.sign == '1':
                val = (-1) * val

        return val
```
